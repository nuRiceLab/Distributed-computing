## Examples on how to display the info from Rucio 

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('rucio_info.json')
```

```python
df
```

```python
plt.figure(figsize=(6, 6))
plt.pie(df['site'].value_counts(), labels=df['site'].value_counts().index, autopct='%1.1f%%', startangle=140)
plt.title('Data collections distribution per RSEs')
plt.savefig('dsets_rse.png')
plt.show()
```

```python
# Group the data by site and sum up the sizes
grouped_df = df.groupby('site')['size'].sum().reset_index()

# Calculate the total size in TB
total_size = round(grouped_df['size'].sum()/1.0e3,2)

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(grouped_df['size'], labels=grouped_df['site'], autopct='%1.1f%%', startangle=140)
plt.title(f'Data percentage per RSEs (Total: {total_size} [TB])')
plt.savefig('dsets_size.png')
plt.show()
```

```python
df_sam = pd.read_json('all_sam.json')
```

```python
df_sam
```

```python
from wordcloud import WordCloud
```

```python
word_dict = {}
for index, row in df_sam.iterrows():
    word_dict[row['dataset']] = row['size']

# Create the WordCloud object with the word_dict as input
wordcloud = WordCloud(width=800, height=400, background_color='white', max_words=40).generate_from_frequencies(word_dict)

# Display the WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('datasets_by_size.png')
plt.show()
```

```python
word_dict = {}
for index, row in df_sam.iterrows():
    word_dict[row['dataset']] = row['n_files']

# Create the WordCloud object with the word_dict as input
wordcloud = WordCloud(width=800, height=400, background_color='white', max_words=40).generate_from_frequencies(word_dict)

# Display the WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('datasets_by_n_files.png')
plt.show()
```

```python
# Group the data by site and sum up the sizes
grouped_df = df_sam.groupby(['dataset','site'])['size'].sum().reset_index()

# Calculate the total size in TB
total_size = round(grouped_df['size'].sum()/1.0e3,2)

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(grouped_df['size'], labels=grouped_df['site'], autopct='%1.1f%%', startangle=140)
plt.title(f'Dataset by size and RSE (Total: {total_size} [TB])')
plt.savefig('dsets_size.png')
plt.show()
```

```python
# Group the data by site and sum up the number of files
grouped_df = df_sam.groupby(['dataset','site'])['n_files'].sum().reset_index()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(grouped_df['n_files'], labels=grouped_df['site'], autopct='%1.1f%%', startangle=140)
plt.title(f'Dataset by n_files and RSE ')
plt.savefig('dsets_size.png')
plt.show()
```
